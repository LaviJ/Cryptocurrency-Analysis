```python
data_retrieved = requests.get(api_endpoint).json()
data_retrieved
```

```python
mining_profitability_data = data_retrieved
```

# Converting the json data into a Dataframe

```python
mining_profitability_data_df = pd.json_normalize(data_retrieved, record_path =['Data'])
mining_profitability_data_df
```

```python
# Replace NaN as '0'
mining_profitability_data_df.isnull().sum().sum()
mining_profitability_data_df.isna().sum()
```

```python
# Drop columns that have null values as those columns are irrelevant
mining_profitability_data_df = mining_profitability_data_df.dropna(axis=1)
mining_profitability_data_df
```

```python
# dropping irrelevant columns
mining_profitability_data_df = mining_profitability_data_df.drop(columns=['Message'])
mining_profitability_data_df.head()
```

# Analyzing Top 10 currencies based on Mining Average Profit Ratio

```python
# Top 10 currencies based on AvgProfitRatio
Top_mining_profitable_currencies_df = mining_profitability_data_df.sort_values(by = ["AvgProfitRatio"], ascending = False)
Top_mining_profitable_currencies_df.head(10)
```

```python
# Top 10 currencies based on Difficulty
Top_mining_difficulty_currencies_df = mining_profitability_data_df.sort_values(by = ["Difficulty"], ascending = False)
Top_mining_difficulty_currencies_df.head(10)
```

# Exporting data to csv and Uploading on MongoDb

```python
# exporting the dataframe to csv
mining_profitability_data_df.to_csv('Mining_profit_difficulty_coinwarz_data.csv')
```

```python
# Read in mongodb server location as client
client = MongoClient(mongo_uri,tlsCAFile=certifi.where())
```

```python
# Find Databases
client.list_database_names()
```

```python
# connecting to the db on MongoDb and creating collection in the db
db = client['Latest_Cryptodata']
collection = db['Mining_profit_difficulty_coinwarz_data']
```

```python
# adding dataframe to mongoDb
mining_profitability_data_df.reset_index(inplace=True)
mining_profitability_data_df_dict = mining_profitability_data_df.to_dict("records")
```

```python
# Insert collection
collection.insert_many(mining_profitability_data_df_dict)
```
